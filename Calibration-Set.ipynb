```python
#####################################################################################################################################
##                                                                                                                                 ## 
##                                                    CALIBRATION SCAN                                                             ##
##                                                                                                                                 ##
#####################################################################################################################################
from scipy import signal
import numpy as np
import pandas as pd
import math as m
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#####################################################################################################################################
##                                                                                                                                 ## 
##                                                    CALIBRATION SCAN                                                             ##
##                                                                                                                                 ##
#####################################################################################################################################

###########################Loading of the data########################################################################################
DataMS = pd.read_csv("C:/Users/renea/Desktop/Manuscript 3 Latex/CLUSTER Experiments/02_February/03feb01_ccl4_34p8amu_step0p05eV_cal_269nA_res20_6cyc_8p2x10m8mbar", sep='\t')
#############################Manage of the data frame##################################################################################
DataMS_arr = DataMS.to_numpy()
MS_arr = DataMS_arr[0:,1:3]

x = MS_arr[0:,0]*(-1) # x-axis, for the case of calibration the measured voltage are inverted, then it should be multiplied by -1 
y = MS_arr[0:,1]       # y-axis

peak_widths = np.arange(0.5, 1.001) # the values 0.5 and 1.001 define the sensibility of the peak detection algorithm
peak_indices = signal.find_peaks_cwt(y, peak_widths) # function for peak detection
peak_count = len(peak_indices) # indexation of the peaks   

plt.plot(x,y, linewidth = 2)   #function to plot the data
#plt.vlines(x[peak_indices], ymin = 0, ymax = max(y),linestyle = "--", linewidth=1.5)
plt.vlines(x[peak_indices[6]], ymin = 0, ymax = max(y),linestyle = "--", linewidth=1.5) # function to add a vertical line in the peak position
plt.xlim(min(x),max(x))

plt.xlabel("Energy (eV)", color = "black", family = "Serif", fontsize = 12) # Label of the x-axis
plt.ylabel("Intensity (c.p.s.)", color = "black", family = "Serif", fontsize = 12) # Label of the y-axis

plt.text(min(x)-0.15*(max(x)-min(x)),1.10*(max(y)),"03feb01_ccl4.png", fontsize = 11) # name of the plot
plt.text(max(x)-0.25*(max(x)-min(x)),1.10*(max(y)),"CCl$_4$", fontsize = 11) # precursor under study
plt.text(max(x)-0.40*(max(x)-min(x)),0.97*max(y), "Correction =          eV", fontsize = 12) # Text
plt.text(max(x)-0.17*(max(x)-min(x)),0.97*max(y), (-1)*float("{0:.3f}".format((x[peak_indices[6]]))), fontsize = 12) # Text
plt.text(max(x)-0.35*(max(x)-min(x)),0.87*max(y), "FWHM = 0.459", fontsize = 12) 

calibre03feb = (x[peak_indices[6]])* (-1) # indexation of the calibration value, it is later read by the another measurements of the day

plt.savefig("C:/Users/renea/Desktop/Manuscript 3 Latex/Version Python/February Corrected/03feb01_ccl4.png", dpi=350, bbox_inches='tight', pad_inches=0.3) # saving of the plot
```
